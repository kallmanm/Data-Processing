### List Comprehension and for loops

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#skapa lista x med fem v채rden 5, 10, 15, 20, 25
x = [5,10,15,20,25]
#skapa tom lista y
y = []
```

```python
#vi vill fylla y med v채rden fr책n x delat med 5
try:
    y = x/5 
    print(y)
    #y = np.array(x)/5
except:
    print('not working')
    
```

```python
# traditional way of for looping
for number in x:
    y.append(number/5)
```

```python
print(y)
```

## List Comprehension

```python
# maybe a better one liner
# ENTER LIST COMPREHENSION
z = [number/5 for number in x]
print(z)
# works with zip() but zip needs to be in a list() ----> list(zip(1,2,3))
```

## Matplotlib lite plot axis modification

```python
# Generera line채r siffer serie med linspace
x = np.linspace(0,20, num=1000)
x
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x, y1, "-g", label="sinusv책g")
plt.plot(x, y2, "-b", label="cosinus")
plt.legend(loc=4)
```

```python
plt.plot(x, y1, "-g", label="sinusv책g")
plt.plot(x, y2, "-b", label="cosinus")
plt.legend(loc=4)
plt.ylim(-3.5,3.5)
plt.xlim(0,20)
```

## BarChart 철verning med matplotlib

```python
#### l채s en .txt och rita den med matplotlib
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

#### skapa en txtfil names.txt med f철ljande namn


James, Johnny, Dennis, Mattias, Sara, Jessica, Robert, Paul, Mr T, Holmes


#### skapa en txtfil salaries.txt med f철ljande data:
0, 100, 200, 500, 1000, 1200, 1800, 1850, 5000, 10000 

```python
salaries = np.fromfile("salaries.txt", dtype=int, sep=",")
names = np.genfromtxt("names.txt", dtype=str,delimiter=",")
```

```python
plt.figure(figsize=(8,4)) # anv채nder figsize f철r att bredda graffen
plt.bar(names,salaries)
```

```python
plt.figure(figsize=(8,4))
plt.bar(names,salaries)
plt.ylabel("L철ner")
#plt.xlabel("Namn")
plt.title("L철nen f철r 10 personer")
```

### lite aritmetik med numpy arrays

```python
print(np.max(salaries), np.min(salaries), np.average(salaries), np.median(salaries))
```

```python
# vi har lite trubbel med grafen
plt.figure(figsize=(8,4))
plt.bar(names[1:8],salaries[1:-2])
plt.ylabel("L철ner")
#plt.xlabel("Namn")
plt.title("L철nen f철r 8 personer")
```

```python
print(np.max(salaries[1:-2]), np.min(salaries[1:-2]), np.average(salaries[1:-2]), np.median(salaries[1:-2]))
```
