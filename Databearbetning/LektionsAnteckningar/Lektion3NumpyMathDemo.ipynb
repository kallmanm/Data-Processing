```python
x = np.linspace(0,10,40) # En fin numpy array med line√§r distribution mellan 0 och 10
```

```python
x b# vi har genererat en x-axel
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y)
```

```python
y2 = np.cos(x)
```

```python
plt.plot(x,y)
plt.plot(x,y2)
plt.legend(['Sin(x)','Cos(x)'])
```

```python
m = np.array([1,3,5,7],'d')
n = np.array([10,5,2,7], 'd')
plt.plot(m,n, 'o')
plt.axis([0,8,0,12])
```

```python
ms = np.linspace(1,7) # skapa j√§mnt f√∂rdelade v√§rden
ns = np.interp(ms,m,n)
plt.plot(ms,ns,'.')
plt.plot(m,n, 'o')
```
