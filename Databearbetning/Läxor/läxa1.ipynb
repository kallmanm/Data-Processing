## Encapsulation & Prime Number Checker

### Version 1
#### Overview
This program checks to see if the user input value is a prime number.

#### Logic
* **get_input()** checks that the user input data is a integer and validates at the same time.
* **evaluate_prime()** does the examines the input number and confirms if it is a prime number or not. Either returning True or False.
* **display_answer()** gives a comand line prompt to the user confirming if input number is prime or not.
* **quit_program()** checks weather to continue running the program or not.
* **run_program()** encapsulates all the above functions to create a logical and cohesive flow.

```python
def get_input():
    '''get_input prompts the user to enter a number so that the program can 
    check weather or not the number enter is a prime number or not.
    Before the value is passed on in the return statement'''
    
    user_input = input("Enter a number to check and see if it's a Prime Number!\n")
    try:
        prime_suspect = int(user_input)
        return prime_suspect
    except ValueError:
        print("Input '{}' is not a valid integer." .format(user_input))
        return 0
    
def evaluate_prime(number=None):
    '''if a number is passed to evaluate_prime it checks weather or not it is a prime number and returns
    a  boolean, True for prime numbers and False for non prime numbers.
    If no number is passed to evaluate_prime it continues without doing an evaluation.'''
    
    #primes have to be bigger than 1.
    if number > 1:
        for i in range(2,number):
            if (number % i) == 0:
                # if modulus remainder is 0 it means that we were able to divide the number we are checking verse 'i'
                # which is a number range between 2 and 'number' meaning it was able to divide with no remainder with
                # a number other than itself. Thus not meeting the requirement of a prime number
                #we then break the loop to 
                
                return False, number
        else:
            return (True, number)
    else:
        return (False, number)
def display_answer(answer):   
    if (answer[0] == True):
        print("{} is a prime number" .format(answer[1]))
    else:
        print("{} is not a prime number." .format(answer[1]))
        
def quit_program(check):
    if(check == "n"):
        return False
    else:
        return True

def run_program():
    program_is_on = True
    while program_is_on:
        display_answer(evaluate_prime(get_input()))
        check_another_number = input("Do you wish to check another number?(y/n)").lower()
        program_is_on = quit_program(check_another_number)
    
run_program()
exit
```
