#### Här får vi medelrating per film genom att använda .groupby() metoden och matar in paramentern den skall sortera enligt 'item id'. Det returnerar åt oss ett pandas objekt som vi kan manipulera. Vi kör tillsist .men() för att få medelvärdet per group.

#### Top 10 filmer som har flera än 40 ratings
1. Först grupperar vi enligt grupp 'item id', sedan filtrera vi datan med en lambda funktion där vi kollar mängden för varje id ska vara vöer 40.

2. Efter det så grupperar vi den nya datan enligt 'item id' och sorterar datan enligt 'rating' och lägger det i storleksordning störst till minst. Därefter splicar vi datan för att enbart spara de top 10 bästa filmerna

3. Till sist returnerar vi datan

```python
df_u_data_top_10_over_40_ratings = df_u_data_sorted_by_item_id.groupby(
    'item id').filter(lambda x: x['rating'].count() > 40)

df_u_data_top_10_over_40_ratings = df_u_data_top_10_over_40_ratings.groupby(
    'item id').mean().sort_values('rating', ascending=False)[:10]

df_u_data_top_10_over_40_ratings
```

#### Skapar en funktion som  filtrerar datan enligt kön och yrken

```python
def most_popular_job_by_gender(gender):
    df_u_user_common_jobs = df_u_user[(df_u_user['gender']== gender)]
    del df_u_user_common_jobs['user id'],df_u_user_common_jobs['gender'],df_u_user_common_jobs['zip code']
    df_u_user_common_jobs = df_u_user_common_jobs.groupby(
        'occupation').count().sort_values('age', ascending=False)
    df_u_user_common_jobs.columns = ['count']
    return df_u_user_common_jobs
```

#### Här använder jag mig av funktionen most_popular_job_by_gender() för att få det vanligaste yrke bland män och kvinnor
